```python
# Import Libraries -------------------------------------------------------------
import pandas as pd
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, facet_wrap

hello!!

# Reading Data -----------------------------------------------------------------
netflix = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/netflix_titles.csv')


# Movies -----------------------------------------------------------------------
movies = netflix[ netflix['type'] == 'Movie']
# movies.head()

ratings = ['G', 'PG', 'PG-13', 'R', 'NC-17']
movies2 = movies[ movies['rating'].isin(ratings) ]
movies2['rating'].value_counts()

ggplot(movies2, aes(x='rating')) + \
    geom_bar(fill='skyblue') + \
    labs(title='Movie Ratings Distribution', x='Rating', y='Count') + \
    theme_minimal()
```

```python
# Tv Shows ---------------------------------------------------------------------
tvshows = netflix[ netflix['type'] == 'TV Show']
# tvshows.head()

tvratings = ['TV-Y', 'TV-Y7', 'TV-G']
shows2 = tvshows[ tvshows['rating'].isin(tvratings) ]
shows2['rating'].value_counts()

ggplot(shows2, aes(x='rating')) + \
    geom_bar(fill='skyblue') + \
    labs(title='Movie Ratings Distribution', x='Rating', y='Count') + \
    theme_minimal()

```
